```python
##removes the duplicates and resets the index.
df = df.drop_duplicates(keep='first').reset_index(drop=True)
df.shape
```

```python
#Checking for outliers (using IQR method for numeric columns only)
print("\n=== Outlier Detection (IQR Method) ===")
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {len(outliers)} potential outliers")
```

```python
## To check the top 10 highest salary outliers.
Q1 = df['salary_in_usd'].quantile(0.25)
Q3 = df['salary_in_usd'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier_rows = df[(df['salary_in_usd'] < lower_bound) | (df['salary_in_usd'] > upper_bound)]
outlier_rows.sort_values(by='salary_in_usd', ascending=False).head(10)

```

```python
## To check the number of potential outliers
# Select numeric columns
numeric_cols = df_1.select_dtypes(include=['int64','float64']).columns

# IQR outlier detection
for col in numeric_cols:
    Q1 = df_1[col].quantile(0.25)
    Q3 = df_1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_1[(df_1[col] < lower_bound) | (df_1[col] > upper_bound)]
    print(f"{col}: {len(outliers)} potential outliers")
```

```python

# H₀ (Null Hypothesis): Higher Generative AI adoption in countries does not correlate with growth in AI/Data job roles.

# H₁ (Alternative Hypothesis): Higher Generative AI adoption in countries does correlate with growth in AI/Data job roles.
    
# Standardizing country fields (strip spaces, unify letter case)

df['company_location'] = df['company_location'].str.strip().str.title()
df_1['Country'] = df_1['Country'].str.strip().str.title()

# checking unique job categories
print(df['job_category'].unique())

ai_categories = [
    'Machine Learning And Ai',
    'Data Science And Research',
    'Data Engineering',
    'Data Analysis',
    'Data Architecture And Modeling',
    'Data Management And Strategy',
    'Data Quality And Operations',
    'Bi And Visualization'
]

df['job_category'] = df['job_category'].astype(str).str.strip().str.title()
ai_data = df[df['job_category'].isin(ai_categories)].copy()

job_counts = ai_data.groupby('company_location').size().reset_index(name='job_count')
job_counts = job_counts.rename(columns={'company_location': 'Country'}).sort_values('job_count', ascending=False)

print("Selected AI/Data rows:", len(ai_data))
print("\nTop 20 countries by AI/Data job_count:")
print(job_counts.head(20).to_string(index=False))
# Checking which countries appear in the GenAI adoption dataset
df_1['Country'] = df_1['Country'].astype(str).str.strip().str.title()

print("Unique countries in GenAI adoption dataset:")
print(df_1['Country'].unique())

print("\nTotal countries:", df_1['Country'].nunique())
# Fixing country name mismatches
df_1['Country'] = df_1['Country'].replace({
    'Usa': 'United States',
    'Uk': 'United Kingdom'
})
adoption_country = df_1.groupby('Country', as_index=False)['Number of Employees Impacted'].sum()

# Merging job counts with adoption data
merged = pd.merge(job_counts, df_1[['Country','Number of Employees Impacted']]), 
merged = pd.merge(job_counts, adoption_country, on='Country', how='inner')

print(merged)
print("Rows:", len(merged))

from scipy.stats import pearsonr, spearmanr
from scipy import stats

x = merged['Number of Employees Impacted']
y = merged['job_count']

# Pearson correlation
pearson_r, pearson_p = pearsonr(x, y)

# Spearman correlation (non-parametric)
spearman_r, spearman_p = spearmanr(x, y)

# Linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print("=== Pearson Correlation ===")
print("r:", pearson_r)
print("p-value:", pearson_p)

print("\n=== Spearman Correlation ===")
print("rho:", spearman_r)
print("p-value:", spearman_p)

print("\n=== Linear Regression ===")
print("slope:", slope)
print("intercept:", intercept)
print("R-squared:", r_value**2)
print("p-value:", p_value)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(merged['Number of Employees Impacted'], 
            merged['job_count'], 
            s=100, alpha=0.7)

# Add labels on each point
for i in range(len(merged)):
    plt.text(merged['Number of Employees Impacted'][i],
             merged['job_count'][i],
             merged['Country'][i],
             fontsize=9)

plt.xlabel("GenAI Adoption (Employees Impacted)")
plt.ylabel("AI/Data Job Count")
plt.title("GenAI Adoption vs AI/Data Job Demand by Country")
plt.grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()

plt.figure(figsize=(12,6))
plt.bar(merged['Country'], merged['job_count'], color='skyblue')
plt.xticks(rotation=45)
plt.ylabel("Job Count")
plt.title("AI/Data Job Count by Country")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,6))
plt.bar(merged['Country'], merged['Number of Employees Impacted'], color='orange')
plt.xticks(rotation=45)
plt.ylabel("Employees Impacted by GenAI")
plt.title("GenAI Adoption by Country")
plt.tight_layout()
plt.show()

import matplotlib.pyplot as plt

countries = merged['Country']
job_counts = merged['job_count']
impacted = merged['Number of Employees Impacted']

plt.figure(figsize=(12,6))

# FIRST line (left axis)
fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(countries, job_counts, color='blue', marker='o', label='Job Count')
ax1.set_ylabel("AI/Data Job Count", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# SECOND line (right axis)
ax2 = ax1.twinx()
ax2.plot(countries, impacted, color='orange', marker='o', label='Employees Impacted')
ax2.set_ylabel("GenAI Employees Impacted", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.xticks(rotation=45)
plt.title("AI/Data Job Count vs GenAI Adoption (Dual Axis Line Chart)")
plt.tight_layout()
plt.show()
```
